#####  Prediction and Observation Methods

```python
# basel station 
sns.set(font_scale = 1.8)
fig, axes = plt.subplots(3,2 ,figsize=(22,18))
fig.suptitle('Basel (Rhine)', fontsize=24, fontweight ="bold")
for index,item in enumerate(data_path): 
  df_data = pd.DataFrame()
  df_data['obs']=pd.DataFrame(item["obs"])
  df_data['pred']=item['pred']
  sb=sns.regplot(ax=axes[int(index/2),index%2],data=df_data, x="pred", y="obs")
  sb.set(xlabel=None)
  sb.set(ylabel=None)

  axes[int(index/2),index%2].set_title(item.get("name"), fontsize=20)
fig.text(0.512, 0.075, 'Predicted streamflow (m/s)', ha='center', fontsize=20)
fig.text(0.512, 0.075, 'Predicted streamflow (m/s)', ha='center', fontsize=20)
fig.text(0.075, 0.5, 'Observed streamflow (m/s)', va='center', rotation='vertical', fontsize=20)
fig.show()


plt.savefig("/content/drive/MyDrive/Thesis-Afshari/figures/scatterplot_basel.png") 
```

# Cumulative distribution curve

```python
# Basel station 
fig, ax = plt.subplots(3,2, sharey = 'all', sharex = 'all', figsize = (26,22))

for index,dt in enumerate(data_path):
  i=index
  #obtain the cumulative frequency of obsered and predicted streamflow
  values, base = np.histogram(dt["pred"], weights=np.ones_like(dt["pred"])/len(dt["pred"]), bins=200)
  cumulative = np.cumsum(values)
  values_o, base_o = np.histogram(dt["obs"], weights=np.ones_like(dt["obs"])/len(dt["obs"]), bins=200)
  cumulative_o = np.cumsum(values_o)

  #plot the results
  ax[i%3,(i>2)*1].plot(cumulative, base[:-1], c='red', label = dt["name"])
  ax[i%3,(i>2)*1].plot(cumulative_o, base_o[:-1], c='black', label = 'obs')
  ax[i%3,(i>2)*1].set_title( dt["name"])
  ax[i%3,(i>2)*1].legend()

#set title and axes labels
fig.text(0.512, 0.075, 'CDF', ha='center', fontsize = 20)
fig.text(0.512, 1 - 0.075, "Basel (Rhine)", ha='center', fontsize = 22, fontweight ="bold")

fig.text(0.075, 0.5, 'Streamflow (m/s)', va='center', rotation='vertical', fontsize = 20)

plt.savefig("/content/drive/MyDrive/Thesis-Afshari/figures/cumulative_Basel.png")
```
